# MERGE de los dos dataframes para luego ser exportado a un .csv

```python
disney_df
```

```python
# Añado una columna adicional para cada dataframe que indique su plataforma.
# De esta manera podemos cargar todos los registros a una única tabla.
disney_df['platform'] = 'Disney'
netflix_df['platform'] = 'Netflix'

# Borro la columna de sentimiento porque no me parecía interesante que esté en la base de datos.
disney_df.drop(columns='description_sentiment', inplace=True)
netflix_df.drop(columns='description_sentiment', inplace=True)


# Reorganizo el show_id, ya que sino muchos quedarían duplicados:

# Calculo el index de partida para el dataframe de Netflix
starting_index = len(disney_df) + 1

# Ajusto los valores de show_id en el dataframe de Netflix
netflix_df['show_id'] = 's' + (starting_index + netflix_df.index).astype(str)

# Se hace el merge de los dataframes
merged_data = pd.concat([disney_df, netflix_df])

# Se hace un reset del index del merged dataframe
merged_data.reset_index(drop=True, inplace=True)
```

```python
duplicates = merged_data.duplicated(subset='show_id')
print(merged_data[duplicates])
```

```python
# Opté por usar solo esta función creada desde el archivo "fill_missing_values.py" ya que la general (fill_data) tardaba
# muchísimo debido a la cantidad de filas. La API es gratuita pero sí es cierto que serían muchas llamadas.
fill_missing_values.fill_country(merged_data)
```

```python
fill_missing_values.fill_release_year(merged_data)
```

```python
# Hallazgos finales: En las filas que tienen "Louis C.K" en el titulo, la duracion de la pelicula estaba en la columna de rating.

# Filtrar los títulos que contienen "Louis C.K"
mask = merged_data['title'].str.contains('Louis C.K')
filtered_data = merged_data[mask]

# Mover los valores de la columna "rating" a la columna "duration"
merged_data.loc[mask, 'duration'] = filtered_data['rating']

# Actualizar los valores de la columna "rating" a None
merged_data.loc[mask, 'rating'] = None
```
